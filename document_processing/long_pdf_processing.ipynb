# Long-form PDF document processing


<br>
For long-form PDF document processing, see company_research folder, which contains:<br>
* document_sourcing.py (.ipynb): example of downloading PDF documents from the web;<br>
* document_processing.py (.ipynb): example of triggering text and quants extraction from PDF documents;<br>
* data_extraction.py (.ipynb): example of extracted data related to specific keyphrases/metrics from documents, and re-structuring that data into a desired form.<br>
