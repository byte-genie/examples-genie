## Taxonomy description


<br>
<p><br>
The classified metrics can be specified in a taxonomy. <br>
Here is a sample of the taxonomy we used for this example: <br>

    {'calc_metric': '(Depreciation for Fixed Assets)', 'symbol': 'DEPFA', 'statement': 'balance-sheet'},<br>
    {'calc_metric': '(Impairment for Invested Properties)', 'symbol': 'IMPINVP', 'statement': 'balance-sheet'},<br>

    {'calc_metric': 'Account Payables', 'symbol': 'ACCPY', 'statement': 'balance-sheet'},<br>

     'statement': 'profit-and-loss-statement'},<br>

    {'calc_metric': '(Other Income)', 'symbol': 'OTHINC', 'statement': 'profit-and-loss-statement'},<br>
    {'calc_metric': 'Bad Debt Written Off / (Written Back)', 'symbol': 'BDWO', 'statement': 'profit-and-loss-statement'}<br>
]<br>
Creating a new financial taxonomy simply requires creating a dataset with `['calc_metric', 'symbol', 'statement']` columns. <br>
Once a new taxonomy is created, it can be saved to ByteGenie database, <br>
and then it can be used in `taxonomise_data` endpoint, to map any data onto that taxonomy. <br>
`taxonomise_data` first classifies each reported metric into one of the statements specified in the taxonomy under <br>
`statement` columns, and then it classifies the metric into one of the metrics specified in `calc_metric` <br>
column in the taxonomy.<br>
The thing to note in creating a taxonomy is that it is better to use long, descriptive names in <br>
`statement` and `calc_metric` columns, instead of acronyms, to avoid classification errors.  <br>
