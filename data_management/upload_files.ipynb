# Upload local files

```python
import os
import base64
import pandas as pd
from utils.byte_genie import ByteGenie
```

## init byte-genie

### init byte-genie in async mode (tasks will run in the background)

```python
bg_async = ByteGenie(
    secrets_file='secrets.json',
    task_mode='async',
    verbose=1,
)
```

### init byte-genie in sync mode (tasks will run in the foreground)

```python
bg_sync = ByteGenie(
    secrets_file='secrets.json',
    task_mode='sync',
    verbose=1,
)
```


<br>
'async' mode is suitable for long-running tasks, so that api calls can be run in the background, <br>
while the rest of the code can continue doing other things.<br>
'sync' mode is suitable for short-lived tasks, where we need some output, before we can move on to anything else.<br>


## Specify the directory containing files to upload

```python
directory = r'/tmp/sample-files'
```

## Get all file names in the directory

```python
file_names = os.listdir(directory)
```

Generate the fileContents and names arrays

```python
contents = []
filenames = []
```

```python
for file_name in file_names:
    file_path = os.path.join(directory, file_name)
    with open(file_path, 'rb') as file:
        content = file.read()
        data = base64.b64encode(content)
        data_with_prefix = "data:;base64," + data.decode('utf-8')
        contents.append(data_with_prefix)
        filenames.append(file_name)
```

## upload files

```python
resp = bg_sync.upload_data(
    contents=contents,
    filenames=filenames,
    username=bg_sync.read_username()
)
```

## check resp

```python
resp.get_status()
```

```python
resp = bg_sync.show_uploads(
    username=bg_sync.read_username()
)
```

### get response data

```python
df_uploads = pd.DataFrame(resp.get_data())
```
